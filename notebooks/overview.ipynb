```python
import mne
import matplotlib.pyplot as plt
```

```python
file_name = "../raw/sample.set"
raw = mne.io.read_raw_eeglab(file_name, preload=True)
raw.plot(n_channels=10, scalings='auto')
```

# E Channel Mappings to Traditional EEG Locations

The E channels in the EGI 128-channel system correspond to traditional EEG electrode locations as follows:

## Key Channel Mappings
- **E36**  Fz (Frontal midline)
- **E104**  Cz (Central midline)
- **E129**  Pz (Parietal midline)
- **E58**  T7/T3 (Left temporal)
- **E96**  T8/T4 (Right temporal)
- **E45**  F7 (Left frontal)
- **E108**  F8 (Right frontal)
- **E70**  O1 (Left occipital)
- **E83**  O2 (Right occipital)
- **E33**  F3 (Left frontal)
- **E124**  F4 (Right frontal)
- **E52**  C3 (Left central)
- **E92**  C4 (Right central)
- **E61**  P3 (Left parietal)
- **E78**  P4 (Right parietal)

## Scalp Region Distribution
- **Frontal region**: E1-E35 (approximately)
- **Central region**: E36-E87 (approximately)
- **Parietal region**: E54-E87 (approximately)
- **Occipital region**: E65-E83 (approximately)
- **Temporal regions**: E45-E58 (left), E96-E108 (right)

> **Note**: EGI numbering doesn't follow anatomical order like the 10-20 system. Each electrode samples brain activity from the underlying cortical areas.

```python
# The E channels represent EGI (Electrical Geodesics Inc.) HydroCel GSN 128-channel system
print("EGI HydroCel GSN 128-Channel System Explanation:")
print("=" * 60)
print("The 'E' channels are from an EGI (Electrical Geodesics Inc.) HydroCel Geodesic Sensor Net.")
print("This is a high-density EEG system with 128 electrodes plus a reference electrode (Cz).")
print("\nKey characteristics:")
print("- E1-E128: Individual electrode positions")
print("- Cz: Central reference electrode")
print("- High-density coverage of the entire scalp")
print("- Geodesic arrangement for optimal spatial sampling")

try:
    if raw.info['dig'] is not None:
        print("\n Montage information found! Plotting electrode locations...")
        fig = raw.plot_sensors(kind='topomap', show_names=True, title='EGI 128-Channel Electrode Positions')
        plt.show()
    else:
        # Load standard EGI montage
        montage = mne.channels.make_standard_montage('GSN-HydroCel-128')
        raw_copy = raw.copy()
        raw_copy.set_montage(montage)
        
        fig = raw_copy.plot_sensors(kind='topomap', show_names=True, title='Standard EGI 128-Channel Layout')
        plt.show()
        
except Exception as e:
    print(f"\nCould not plot sensors: {e}")
    print("Will show standard layout information instead.")
```
