```python
import mne
import os

despicable_me = []
fun_with_fractals = []
wimpy_kid = []
the_present = []

raw_dir = os.path.join('..', 'raw')

n = 1
while True:
    fname = os.path.join(raw_dir, f'despicable_me{n}.set')
    if not os.path.exists(fname):
        print(f'No file found: {fname} — stopping (last checked {n}).')
        break
        
    print(f'Processing: {fname}')
    raw = mne.io.read_raw_eeglab(fname, preload=True)

    despicable_me.append(raw)
    n += 1

n = 1
while True:
    fname = os.path.join(raw_dir, f'fun_with_fractals{n}.set')
    if not os.path.exists(fname):
        print(f'No file found: {fname} — stopping (last checked {n}).')
        break
        
    print(f'Processing: {fname}')
    raw = mne.io.read_raw_eeglab(fname, preload=True)

    fun_with_fractals.append(raw)
    n += 1

n = 1
while True:
    fname = os.path.join(raw_dir, f'wimpy_kid{n}.set')
    if not os.path.exists(fname):
        print(f'No file found: {fname} — stopping (last checked {n}).')
        break
        
    print(f'Processing: {fname}')
    raw = mne.io.read_raw_eeglab(fname, preload=True)

    wimpy_kid.append(raw)
    n += 1

n = 1
while True:
    fname = os.path.join(raw_dir, f'the_present{n}.set')
    if not os.path.exists(fname):
        print(f'No file found: {fname} — stopping (last checked {n}).')
        break
        
    print(f'Processing: {fname}')
    raw = mne.io.read_raw_eeglab(fname, preload=True)

    the_present.append(raw)
    n += 1

all_recordings = {
    'Despicable Me': despicable_me,
    'Fun with Fractals': fun_with_fractals,
    'Wimpy Kid': wimpy_kid,
    'The Present': the_present
}

all_recordings
```

```python
# combined plotting for a selectable naturalistic stimulus dataset
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy.stats import zscore
import numpy as np

chan_name = 120

# choose which dataset to plot: one of 'despicable_me', 'fun_with_fractals', 'wimpy_kid', 'the_present'
data_set = 'the_present'

dataset_map = {
    'despicable_me': despicable_me,
    'fun_with_fractals': fun_with_fractals,
    'wimpy_kid': wimpy_kid,
    'the_present': the_present,
}

if data_set not in dataset_map:
    raise ValueError(f"Unknown dataset '{data_set}'. Choose one of: {list(dataset_map.keys())}")

data_list = dataset_map[data_set]

def format_mmss(x, pos):
    minutes = int(x // 60)
    seconds = int(x % 60)
    return f"{minutes}:{seconds:02d}"

if len(data_list) == 0:
    print('selected dataset is empty — run the loader cell (Cell 1) first to populate recordings.')
else:
    plt.figure(figsize=(12, 4))
    ax = plt.gca()

    cmap = plt.get_cmap('tab10')
    n_recs = len(data_list)

    for i, raw in enumerate(data_list):
        data = raw.get_data(picks=[chan_name])[0]
        sfreq = raw.info['sfreq']

        # use the full recording (no trimming)
        data_trim = data
        ts = np.arange(len(data_trim)) / sfreq

        try:
            from biosppy.signals import eeg
            out = eeg.eeg(signal=data_trim, sampling_rate=sfreq, show=False)
            trace = zscore(out['filtered'])
        except Exception:
            trace = zscore(data_trim)

        color = cmap(i % cmap.N)
        ax.plot(ts, trace, label=f'{data_set}_{i+1}', color=color, linewidth=0.9)

    ax.set_xlabel('Time (mm:ss)')
    ax.set_ylabel('Normalized Amplitude (z-score)')
    ax.set_title(f'Combined traces — channel {chan_name} (dataset: {data_set})')
    ax.grid(True, alpha=0.25)

    # Bottom axis formatting: minutes (major) and 10s minor ticks
    ax.xaxis.set_major_locator(ticker.MultipleLocator(60))
    ax.xaxis.set_minor_locator(ticker.MultipleLocator(10))
    ax.xaxis.set_major_formatter(ticker.FuncFormatter(format_mmss))

    ax_top = ax.twiny()
    ax_top.set_xlim(ax.get_xlim())
    ax_top.xaxis.set_major_locator(ticker.MultipleLocator(30))
    ax_top.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f"{int(x)}s"))
    ax_top.set_xlabel('Seconds')

    ax.legend(loc='upper right', fontsize='small', ncol=2)
    plt.tight_layout()
    plt.show()
```
